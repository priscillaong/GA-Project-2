# 1. Problem Statement

We are the Data team working at NexRes which is a property portal trusted for its extensive listings and comprehensive market information on residential property in Iowa. <br><br>
The aim of this project is to develop a Regression model that can predict to at least 80% accuracy of Iowa property prices by analysing and narrowing down a list of 80 housing features. <br><br>
This would help agents registered under NexRes to predict property sale prices and use this predictor to advise clients to optimise property for sale and maximise profits of their property. <br><br>
It is imperative for us to develop this model to maintain our status in the growing industry and empower our agents to continue providing top value to our clients, who rely on us to secure the best deal.

# 4. Data Cleaning

```python
#Import Libraries
import pandas as pd

#Suppress Warnings
import warnings
warnings.filterwarnings('ignore')

#Adjust the max columns displayed in the df
pd.set_option("display.max_columns", 85)
```

Import train file into df for data cleaning

```python
df_train = pd.read_csv("../Dataset/train.csv")
```

Taking a look at the general shape of the df

```python
df_train.shape
```

Taking a look at the first few records of df

```python
df_train.head()
```

Further info on the columns and their dtypes

```python
df_train.info()
```

## Check for Null Values

```python
df_train[df_train.columns[df_train.isnull().any()]].isnull().sum()
```

Null values as a percentage of total number of records

```python
df_train[df_train.columns[df_train.isnull().any()]].isnull().sum() * 100 / df_train.shape[0]
```

Remove columns where >80% of column is null and save as new df: df_train_clean

```python
df_train_clean = df_train.drop(labels=['Alley', 'Pool QC', 'Fence', 'Misc Feature'], axis = 1)
```

## Impute Missing Values

Take a look at Lot frontage which has null values and its lot area

```python
df_train_clean[df_train_clean["Lot Frontage"].isnull()].head()
```

To replace Lot Frontage with its median (should not be null since there is a value for lot area)

```python
median = df_train_clean['Lot Frontage'].median()
df_train_clean['Lot Frontage'] = df_train_clean['Lot Frontage'].fillna(median)
```

Mas Vnr Type and Mas Vnr Area has 22 null values, to fill the null values with None and 0

```python
df_train_clean['Mas Vnr Type'] = df_train_clean['Mas Vnr Type'].fillna("None")
df_train_clean['Mas Vnr Area'] = df_train_clean['Mas Vnr Area'].fillna(0.0)
```

To address the null values in the basement variables: <br>
To remove rows 1456, 1547 and 1997 as only basement exposure info is missing in these rows

```python
df_train_clean[["Bsmt Qual", "Bsmt Cond", "Bsmt Exposure", "BsmtFin Type 1", "BsmtFin SF 1", "BsmtFin Type 2", "BsmtFin SF 2", "Bsmt Unf SF", "Total Bsmt SF"]].iloc[[1456, 1547, 1997]]
```

To address the null values in the basement variables: <br>
To remove row 1147 as only BsmtFin Type 2 info is missing in this row

```python
df_train_clean[["Bsmt Qual", "Bsmt Cond", "Bsmt Exposure", "BsmtFin Type 1", "BsmtFin SF 1", "BsmtFin Type 2", "BsmtFin SF 2", "Bsmt Unf SF", "Total Bsmt SF"]].iloc[[1147]]
```

Drop rows 1147, 1456, 1547 and 1997

```python
df_train_clean = df_train_clean.drop(df_train_clean.index[[1147, 1456, 1547, 1997]])
```

To reset index after dropping the rows

```python
df_train_clean = df_train_clean.reset_index(drop = True)
```

For basement variables, to replace null numerical variables to 0.0 and categorial variables to No/No Basement

```python
df_train_clean['Bsmt Qual'] = df_train_clean['Bsmt Qual'].fillna("No Basement")
df_train_clean['Bsmt Cond'] = df_train_clean['Bsmt Cond'].fillna("No Basement")
df_train_clean['Bsmt Exposure'] = df_train_clean['Bsmt Exposure'].fillna("No")
df_train_clean['BsmtFin Type 1'] = df_train_clean['BsmtFin Type 1'].fillna("No Basement")
df_train_clean['BsmtFin Type 2'] = df_train_clean['BsmtFin Type 2'].fillna("No Basement")
df_train_clean['BsmtFin SF 1'] = df_train_clean['BsmtFin SF 1'].fillna(0.0)
df_train_clean['BsmtFin SF 2'] = df_train_clean['BsmtFin SF 2'].fillna(0.0)
df_train_clean['Bsmt Unf SF'] = df_train_clean['Bsmt Unf SF'].fillna(0.0)
df_train_clean['Total Bsmt SF'] = df_train_clean['Total Bsmt SF'].fillna(0.0)
df_train_clean['Bsmt Full Bath'] = df_train_clean['Bsmt Full Bath'].fillna(0.0)
df_train_clean['Bsmt Half Bath'] = df_train_clean['Bsmt Half Bath'].fillna(0.0)
```

For fireplace qu, fireplaces number is 0, hence to replace null with No Fireplace

```python
df_train_clean[df_train_clean["Fireplace Qu"].isnull()].head()
```

Replace null values in fireplace qu to No Fireplace

```python
df_train_clean['Fireplace Qu'] = df_train_clean['Fireplace Qu'].fillna("No Fireplace")
```

Take a look at garage variables which have null values

```python
df_train_clean[df_train_clean["Garage Yr Blt"].isnull()].head()
```

To remove row 1709 as garage type info is found but other garage info is missing (null values)

```python
df_train_clean[["Garage Type", "Garage Yr Blt", "Garage Finish", "Garage Cars", "Garage Area", "Garage Qual", "Garage Cond"]].iloc[[1709]]
```

```python
df_train_clean = df_train_clean.drop([df_train_clean.index[1709]])
```

Reset Index after dropping row

```python
df_train_clean = df_train_clean.reset_index(drop = True)
```

For garage variables, to replace null numerical variables to 0.0 and categorial variables to No Basement

```python
df_train_clean['Garage Yr Blt'] = df_train_clean['Garage Yr Blt'].fillna(0.0)
df_train_clean['Garage Type'] = df_train_clean['Garage Type'].fillna("No Garage")
df_train_clean['Garage Finish'] = df_train_clean['Garage Finish'].fillna("No Garage")
df_train_clean['Garage Qual'] = df_train_clean['Garage Qual'].fillna("No Garage")
df_train_clean['Garage Cond'] = df_train_clean['Garage Cond'].fillna("No Garage")
```

All null values have been dealt with! :)

```python
df_train_clean[df_train_clean.columns[df_train_clean.isnull().any()]].isnull().sum()
```

## Check for Duplicates

To check for duplicates and remove if any

```python
df_train_clean.duplicated().sum()
```

## Rename Column Names

```python
def rename(dataframe):
    dataframe.columns = dataframe.columns.str.lower().str.replace(" ", "_")
    print(dataframe.columns)

rename(df_train_clean)
```

```python
#rename some other columns with numbers at the front and other symbols in the feature names

df_train_clean.rename(columns={
    '1st_flr_sf': 'first_flr_sf',
    '2nd_flr_sf': 'second_flr_sf',
    '3ssn_porch': 'three_ssn_porch',
    'year_remod/add': 'year_remod_add'
}, inplace=True)
```

## Save Clean Train File as New csv File

All null values/duplicate data have been dealt with and columns renamed

```python
df_train_clean.info()
```

Save clean df_train file as a new csv file
